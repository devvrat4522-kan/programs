```python
def gener(x):
    s=""
    if len(x) == 0:
        return None
    else:
        a="@".join(x.split("'name':"))
        for i in range(len(a)):
            if a[i]=="@":
                for j in range(i,len(a)):
                    if a[j] != '}':
                        s+=a[j]
                    else:
                        s+='_'
                        break
                    
        return s.replace("@","").replace("'","").replace("_","")[:-1]
            
```

```python
df.genres.apply(gener)
```
