```python
import pandas as pd
dgh=pd.read_csv("movies_metadata.csv")
dgh
```

```python
def gener(x):
    s=""
    if len(x) == 0:
        return None
    else:
        a="@".join(x.split("'name':"))
        for i in range(len(a)):
            if a[i]=="@":
                for j in range(i,len(a)):
                    if a[j] != '}':
                        s+=a[j]
                    else:
                        s+='_'
                        break
                    
        return s.replace("@","").replace("'","").replace("_","")
```

```python
dgh.genres.apply(gener)
```

```python
dgh.genres=dgh.genres.apply(gener)
```

```python
dg=pd.read_csv("movies_metadata.csv")
dg.spoken_languages
```

```python
dg.spoken_languages.dropna(inplace=True)
```

```python
def lang(x):
    try:
        return x.split("'name':")[-1]
    except:
        return x
```

```python
dg.spoken_languages=dg.spoken_languages.apply(lang)
dg.spoken_languages=dg.spoken_languages.apply(lambda x:'no language' if x=='[]' else x)
dg
```

```python
def manag(x):
    try:                                                                            # text cleaning
        return x.split("'name':")[-1].split("}")[0].replace("'","")
    except:
        return x
```

```python
dg.genres.apply(manag)
```

```python
dg.genres=dg.genres.apply(manag)
```
