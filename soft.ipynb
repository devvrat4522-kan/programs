```python
import numpy as np
import matplotlib.pyplot as plt
```

Triangular Membership Function

```python
def triangular_membership(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)
```

```python
x_values = np.linspace(0, 10, 100)
print(x_values)
```

```python
a = 2
b = 5
c = 8
membership_values = [triangular_membership(x, a, b, c) for x in x_values]
print(membership_values)
```

```python
plt.plot(x_values,membership_values)
plt.ylabel('membership_values')
plt.xlabel('X-axis')
plt.title('Triangular Membership function')
plt.show()
```

Gaussian Membership Function


```python
def gaussian_membership(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma)**2)
```

```python
x_values1 = np.linspace(0, 10, 100)
mean = 5
sigma = 2
membership_values1 = [gaussian_membership(x, mean, sigma) for x in x_values1]
```

```python
plt.plot(x_values1,membership_values1)
plt.ylabel('membership_values')
plt.xlabel('X-axis')
plt.title('Triangular Membership function')
plt.show()
```

Fuzzy set operation


```python
def triangular_membership(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)
    

x_values = np.linspace(0, 10, 100)
```

```python
# Define parameters for two triangular membership functions
a1, b1, c1 = 2, 5, 8
a2, b2, c2 = 4, 7, 9

# first membership function
membership_values1 = [triangular_membership(x, a1, b1, c1) for x in x_values]

# second membership function
membership_values2 = [triangular_membership(x, a2, b2, c2) for x in x_values]
```

```python
# Fuzzy set union (maximum operator)
union_values = np.maximum(membership_values1, membership_values2)

# # Fuzzy set complement
# complement_values1 = 1 - membership_values1
# complement_values2 = 1 - membership_values2
```

```python
# Plot the original membership functions
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x_values, membership_values1, label='A')
plt.title('Membership Function A')

plt.subplot(2, 2, 2)
plt.plot(x_values, membership_values2, label='B')
plt.title('Membership Function B')

# Plot the union of the two fuzzy sets
plt.subplot(2, 2, 3)
plt.plot(x_values, union_values, label='A ∪ B', color='red')
plt.title('Fuzzy Set Union (A ∪ B)')
```

```python
# Fuzzy set intersection (minimum operator)
intersection_values = np.minimum(membership_values1, membership_values2)

# Plot the intersection of the two fuzzy sets

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 4)
plt.plot(x_values, intersection_values, label='A ∩ B', color='green')
plt.title('Fuzzy Set Intersection (A ∩ B)')

plt.tight_layout()
plt.show()
```

Composition of Fuzzy Relation

```python
A = np.array([[0.8, 0.2, 0.4],
              [0.5, 0.6, 0.7],
              [0.1, 0.9, 0.3]])

B = np.array([[0.7, 0.3, 0.5],
              [0.2, 0.8, 0.6],
              [0.4, 0.5, 0.9]])

def max_min_composition(A, B):
    rows_A, cols_A = A.shape
    rows_B, cols_B = B.shape

    result = np.zeros((rows_A, cols_B))

    for i in range(rows_A):
        for j in range(cols_B):
            max_min_values = [min(A[i, k], B[k, j]) for k in range(cols_A)]
            result[i, j] = max(max_min_values)

    return result

C = max_min_composition(A, B)

# Display the results
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMax-Min Composition (A o B):")
print(C)
```

```python
import skfuzzy as fuzz

# Generate universe variables
x = np.arange(0, 10, 1)
print(x)

# Generate fuzzy membership functions
# Triangular membership function
triangular_mf = fuzz.trimf(x, [2, 5, 8])

# Trapezoidal membership function
trapezoidal_mf = fuzz.trapmf(x, [2, 4, 6, 9])

# Gaussian membership function
gaussian_mf = fuzz.gaussmf(x, 5, 2)

# Sigmoid membership function
sigmoid_mf = fuzz.sigmf(x, 3, 2)

# Plot membership functions
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.plot(x, triangular_mf, label='Triangular')
plt.figure(figsize=(12,8))
plt.subplot(2,2,2)
plt.plot(x, trapezoidal_mf, label='Trapezoidal')
plt.show()

plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
plt.plot(x, gaussian_mf, label='Gaussian')
plt.show()

plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
plt.plot(x, sigmoid_mf, label='Sigmoid')
plt.show()
```

```python
from skfuzzy import control as ctrl

# Create input and output variables
temperature = ctrl.Antecedent(np.arange(0, 101, 1), 'temperature')
humidity = ctrl.Antecedent(np.arange(0, 101, 1), 'humidity')
fan_speed = ctrl.Consequent(np.arange(0, 101, 1), 'fan_speed')

# Define fuzzy sets for input variables
temperature['low'] = fuzz.trimf(temperature.universe, [0, 0, 50])
temperature['medium'] = fuzz.trimf(temperature.universe, [0, 50, 100])
temperature['high'] = fuzz.trimf(temperature.universe, [50, 100, 100])

humidity['low'] = fuzz.trimf(humidity.universe, [0, 0, 50])
humidity['medium'] = fuzz.trimf(humidity.universe, [0, 50, 100])
humidity['high'] = fuzz.trimf(humidity.universe, [50, 100, 100])

# Define fuzzy sets for output variable
fan_speed['low'] = fuzz.trimf(fan_speed.universe, [0, 0, 50])
fan_speed['medium'] = fuzz.trimf(fan_speed.universe, [0, 50, 100])
fan_speed['high'] = fuzz.trimf(fan_speed.universe, [50, 100, 100])

# Create fuzzy rules
rule1 = ctrl.Rule(temperature['low'] & humidity['low'], fan_speed['low'])
rule2 = ctrl.Rule(temperature['medium'] & humidity['medium'], fan_speed['medium'])
rule3 = ctrl.Rule(temperature['high'] | humidity['high'], fan_speed['high'])

# Create control system
fan_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])

# Create simulation
fan_simulation = ctrl.ControlSystemSimulation(fan_ctrl)

# Input values
fan_simulation.input['temperature'] = 30
fan_simulation.input['humidity'] = 70

# Compute the result
fan_simulation.compute()

# Output value
print("Fan Speed:", fan_simulation.output['fan_speed'])

# Visualize the fuzzy membership functions and rules
temperature.view()
humidity.view()
fan_speed.view()
rule1.view()
rule2.view()
rule3.view()

# Show plots
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Create Antecedent/Consequent objects representing input/output variables
temperature = ctrl.Antecedent(np.arange(0, 101, 1), 'temperature')
rate_of_change = ctrl.Antecedent(np.arange(-5, 6, 1), 'rate_of_change')
heater_power = ctrl.Consequent(np.arange(0, 101, 1), 'heater_power')

# Define fuzzy sets for input variables
temperature['cold'] = fuzz.trimf(temperature.universe, [0, 0, 50])
temperature['comfortable'] = fuzz.trimf(temperature.universe, [0, 50, 100])
temperature['hot'] = fuzz.trimf(temperature.universe, [50, 100, 100])

rate_of_change['decreasing'] = fuzz.trimf(rate_of_change.universe, [-5, -5, 0])
rate_of_change['stable'] = fuzz.trimf(rate_of_change.universe, [-5, 0, 5])
rate_of_change['increasing'] = fuzz.trimf(rate_of_change.universe, [0, 5, 5])

# Define fuzzy sets for output variable
heater_power['low'] = fuzz.trimf(heater_power.universe, [0, 0, 50])
heater_power['medium'] = fuzz.trimf(heater_power.universe, [0, 50, 100])
heater_power['high'] = fuzz.trimf(heater_power.universe, [50, 100, 100])

# Define rules
rule1 = ctrl.Rule(temperature['cold'] & rate_of_change['increasing'], heater_power['high'])
rule2 = ctrl.Rule(temperature['cold'] & rate_of_change['stable'], heater_power['medium'])
rule3 = ctrl.Rule(temperature['comfortable'], heater_power['medium'])
rule4 = ctrl.Rule(temperature['hot'] & rate_of_change['decreasing'], heater_power['low'])

# Create control system
heater_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])

# Create simulation
heater_simulation = ctrl.ControlSystemSimulation(heater_ctrl)

# Input values
heater_simulation.input['temperature'] = 75
heater_simulation.input['rate_of_change'] = 2

# Compute the result
heater_simulation.compute()

# Output value
print("Heater Power:", heater_simulation.output['heater_power'])

# Visualize the fuzzy membership functions and rules
temperature.view()
rate_of_change.view()
heater_power.view()
rule1.view()
rule2.view()
rule3.view()
rule4.view()

# Show plots
plt.show()
```
